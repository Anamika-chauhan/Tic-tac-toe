# TIC-TAC TOE GAME

```python
# To display the tic tac toe board
def print_board(a):
    print("",a[1]," │",a[2]," | ",a[3]," ")
    print("----|----|----")
    print("",a[4]," │",a[5]," │ ",a[6]," ")
    print("----|----|----")
    print("",a[7]," │",a[8]," │ ",a[9]," ")
    
# Some intructions for the game
def show_instructions():
    print("\n\n*********************WELCOME TO TIC TAC TOE*****************\n\n")
    print_board(pos)
    
    print()
    player[0] = input("Player 1 name is : ")
    player[1] = input("Player 2 name is : ")
      
    print("\n<---------------[INSTRUCTIONS]------------------>\n")
      
    print("*", player[0], "will be using x")
    print("*", player[1], "will be using o")
    print("*", player[0], "will start the game")
    print("* Positions on board are like this ...\n")
      
    print(" 1   |   2   |   3 ")
    print(".....|.......|.....")
    print(" 4   |   5   |   6 ")
    print(".....|.......|.....")
    print(" 7   |   8   |   9 ")
      
    print("\nPress Y/y to start the game!")
    instructions = input()
    return instructions
        
    print("\n<------------------------------------------------>\n")
      
# #To start the Game
def start_game():
    turn = 0
    for i in range(9):
        if turn % 2 == 0:
            print("\n",player[0], "'s turn")
            ps = int(input("Enter the position..."))
            
            while pos[ps] == 'x' or pos[ps] == 'o':
                print("\nInvalid Entry place already occupied :(\n")
                ps = int(input("Enter another position..."))

            a = 'x'
            pos[ps] = a
            print_board(pos)
            winner = check_win(a)
            if winner == "nobody":
                turn = 1;
                continue
            else:
                print("\n\n", player[0], "You win :)")
                break
        else:
            print("\n",player[1], "'s turn")
            ps = int(input("Enter the position..."))

            while pos[ps] == 'x' or pos[ps] == 'o':
                print("\nInvalid Entry place already occupied :(\n")
                ps = int(input("Enter another position..."))

            a = 'o'
            pos[ps] = a
            print_board(pos)
            winner = check_win(a)
            if winner =="nobody":
                turn = 0
                continue
            else:
                print("\n\n", player[1], "You win :)")
                break
                
    else:
        print("\nGame drawn\n")
            
# Check for winner 
def check_win(a):
    for i in winning_conditions:
        if (pos[i[0]], pos[i[1]], pos[i[2]]) == (a, a, a):
            winner = player[0]
            break
        elif (pos[i[0]], pos[i[1]], pos[i[2]]) == (a, a, a):
            winner = player[1]
            break
    else:
        winner = "nobody"
    return winner


# Code starts from here (MAIN CODE)
  
pos = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
player = [' ', ' ']
winning_conditions = [(1, 2, 3), (4, 5, 6), (7, 8, 9), (1, 4, 7),
(2, 5, 8), (3, 6, 9), (1, 5, 9), (3, 5, 7)]
instructions = show_instructions()
if(instructions == 'Y' or instructions == 'y'):
    start_game()
else:
    print("Invalid Entry")
```
